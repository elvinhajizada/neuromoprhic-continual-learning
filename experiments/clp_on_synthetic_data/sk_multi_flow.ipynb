```python
%matplotlib notebook
from skmultiflow.data import RandomRBFGenerator
from skmultiflow.bayes import NaiveBayes
from skmultiflow.prototype import RobustSoftLearningVectorQuantization
from skmultiflow.evaluation import EvaluatePrequential
```

```python
# 1. Create a stream
stream = RandomRBFGenerator(n_classes=10, n_features=10, n_centroids=10)
# 2. Instantiate the classifier
nb = NaiveBayes()
lvq = RobustSoftLearningVectorQuantization(prototypes_per_class=1,  sigma=2.5)
# 3. Setup the evaluator
evaluator = EvaluatePrequential(show_plot=True,
                                pretrain_size=100,
                                n_wait = 100,
                                max_samples=2000,
                                metrics=['accuracy'])
# 4. Run evaluation
evaluator.evaluate(stream=stream, model=[nb,lvq])
```
