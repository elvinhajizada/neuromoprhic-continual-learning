```python
%matplotlib notebook
from skmultiflow.data import RandomRBFGenerator
from skmultiflow.bayes import NaiveBayes
from skmultiflow.prototype import RobustSoftLearningVectorQuantization
from skmultiflow.evaluation import EvaluatePrequential
from skmultiflow.neural_networks import PerceptronMask
from skmultiflow.trees import HoeffdingTreeClassifier
from skmultiflow.lazy import KNNClassifier

import importlib

import clp.sk_clp
from clp.sk_clp import CLP
```

```python
importlib.reload(clp.sk_clp)
from clp.sk_clp import CLP
# 1. Create a stream
n_classes=100
stream = RandomRBFGenerator(n_classes=n_classes, n_features=1000, n_centroids=400)
# 2. Instantiate the classifier
nb = NaiveBayes()
lvq = RobustSoftLearningVectorQuantization(prototypes_per_class=1,  sigma=2.5)
perceptron = PerceptronMask()
ht = HoeffdingTreeClassifier()
knn = KNNClassifier()
# 3. Setup the evaluator
evaluator = EvaluatePrequential(show_plot=True,
                                pretrain_size=1,
                                n_wait = 50,
                                max_samples=2800,
                                metrics=['accuracy', 'running_time'])

clp_classifier = CLP(sim_metric='dot_product',
                    n_protos=1000,
                    num_classes=n_classes,
                    alpha_init=0.633, 
                    sim_th_init=0.442, 
                    max_allowed_mistakes=6, 
                    k_hit=1.357,
                    k_miss=0.016,
                    tau_sim_th_pos=100, 
                    tau_sim_th_neg=126,
                    k_sim_th_pos=0.770,
                    k_sim_th_neg=1.1,
                    device='cpu', 
                    verbose=0)
# 4. Run evaluation
evaluator.evaluate(stream=stream, model=[nb,knn,clp_classifier])
```
