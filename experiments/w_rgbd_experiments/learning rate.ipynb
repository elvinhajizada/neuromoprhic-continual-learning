```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x, c1=1, c2=10):
    c1 = 1
    c2 = 10
    return 1 - 1/(1+np.exp(-c1*(x-c2)))
```

```python
def sig_hat(x):
    c1 = 1
    c2 = 10
    return -(c1*np.exp(c1*(c2 + x)))/(np.exp(c1*c2) + np.exp(c1*x))**2
```

```python
x = np.arange(0,20,0.2)
y = sig_hat(x)
```

```python
def gaussian_drv(x, mu=10, sig=2.5):

    return -((x - mu)*np.exp(-(x - mu)**2/(2*sig**2)))/(np.sqrt(2*np.pi)*sig**3)
```

```python
y = gaussian_drv(x)
```

```python
plt.plot(x,y)
plt.show()
```
